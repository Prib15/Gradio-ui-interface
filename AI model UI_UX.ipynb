```python
import gradio as gr
import numpy as np
from PIL import Image

IMG_SIZE = (180, 180)
CLASS_NAMES = {0: "Normal", 1: "Pneumonia"}

def preprocess_xray(img):
    img = img.convert("RGB").resize(IMG_SIZE)
    return np.expand_dims(np.array(img)/255.0, 0)

def predict_xray(img):
    x = preprocess_xray(img)
    prob = model.predict(x)[0][0]
    label = 1 if prob >= 0.5 else 0
    conf = prob if label == 1 else 1 - prob
    return f"{CLASS_NAMES[label]}  (confidence {conf:.2%})"

gr.Interface(
    fn=predict_xray,
    inputs=gr.Image(type="pil", label="Chest X-ray"),
    outputs=gr.Text(label="Prediction"),
    title="Pneumonia Detector",
    description="Upload a chest X-ray image. The CNN will predict if it shows signs of pneumonia.",
    allow_flagging="never"
).launch(share=True)
```
